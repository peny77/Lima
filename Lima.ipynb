## a) How many calories does the average McDonald's value meal contain?

> Berapa rata² Kalori yang terkandung dalam menu² makanan McDonald

```python
f, ax = plt.subplots(1, 2, figsize=(25, 15))
mean_kalori = {}
mean_kolesterol = {}

# Rata-rata Kalori
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Calories'].mean())
    mean_kalori[category] = mean
pd.Series(mean_kalori).sort_values(ascending=False).plot.barh(width=0.8, color=sns.color_palette('winter',10), ax=ax[0])
ax[0].set_title("Mean Calories of Different Categories")

# Rata-rata Kolesterol
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Cholesterol'].mean())
    mean_kolesterol[category] = mean
pd.Series(mean_kolesterol).sort_values(ascending=False).plot.barh(width=0.8, color=sns.color_palette('winter',10), ax=ax[1])
ax[1].set_title("Mean Cholesterol of Different Categories");
```

## b) How much do beverages, like soda or coffee, contribute to the overall caloric intake? 

> Seberapa banyak minuman seperti soda/kopi berkontribusi pada asupan kalori secara keseluruhan

```python
label = df['Category'].unique()
```

```python
category = df.groupby(['Category']).mean()
```

```python
calories_mean = category.Calories.values / 2000
```

### Mean Calories - Category

```python
# Pie chart
labels = label
sizes = calories_mean
explode = (0.05, 0.01, 0.01, 0.01, 0.01, 0.01, 0.1, 0.01, 0.01)

fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
         startangle=90)
ax1.axis('equal')

plt.show()
```

### Beverages

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = label[6:9]
sizes = calories_mean[6:9]
explode = (0.1, 0.01, 0.01)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
         startangle=45)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.barh(labels,sizes)

legend = ax.legend(loc='best', shadow=False, fontsize='medium')
legend.get_frame().set_facecolor('C0')
plt.show()
```
