```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(font_scale=2)
plt.figure(figsize=(8, 5))
sns.countplot(y='Category', data=df, label="Count")
```

## a) How many calories does the average McDonald's value meal contain?

> Berapa rataÂ² Kalori yang terkandung dalam menuÂ² makanan McDonald?*Solusi:* cari rataÂ² kalori tiap kategori, visualisasi dengan Bar chart

```python
f, ax = plt.subplots(1, 2, figsize=(25, 15))
mean_kalori = {}
mean_kolesterol = {}

# Rata-rata Kalori
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Calories'].mean())
    mean_kalori[category] = mean
pd.Series(mean_kalori).sort_values(ascending=False).plot.barh(width=0.8, color=sns.color_palette('winter',10), ax=ax[0])
ax[0].set_title("Mean Calories of Different Categories")

# Rata-rata Kolesterol
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Cholesterol'].mean())
    mean_kolesterol[category] = mean
pd.Series(mean_kolesterol).sort_values(ascending=False).plot.barh(width=0.8, color=sns.color_palette('winter',10), ax=ax[1])
ax[1].set_title("Mean Cholesterol of Different Categories");
```

```python
f, ax = plt.subplots(1, 2, figsize=(25, 15))
mean_sodium = {}
mean_carbohydrates = {}
mean_sugars = {}
mean_protein = {}

# Rata-rata Sodium
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Sodium'].mean())
    mean_sodium[category] = mean
pd.Series(mean_sodium).sort_values(ascending=False).plot.barh(width=0.8, color=sns.color_palette('winter',10), ax=ax[0])
ax[0].set_title("Mean Sodium of Different Categories")

# Rata-rata Carbohydrates
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Carbohydrates'].mean())
    mean_carbohydrates[category] = mean
pd.Series(mean_carbohydrates).sort_values(ascending=False).plot.barh(width=0.8, color=sns.color_palette('winter',10), ax=ax[1])
ax[1].set_title("Mean Carbohydrates of Different Categories");
```

```python
f, ax = plt.subplots(1, 2, figsize=(25, 15))
mean_sugars = {}
mean_protein = {}

# Rata-rata Sugars
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Sugars'].mean())
    mean_sugars[category] = mean
pd.Series(mean_sugars).sort_values(ascending=False).plot.barh(width=0.8, color=sns.color_palette('winter',10), ax=ax[0])
ax[0].set_title("Mean Sugars of Different Categories")

# Rata-rata Protein
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Protein'].mean())
    mean_protein[category] = mean
pd.Series(mean_protein).sort_values(ascending=False).plot.barh(width=0.8, color=sns.color_palette('winter',10), ax=ax[1])
ax[1].set_title("Mean Protein of Different Categories");
```
